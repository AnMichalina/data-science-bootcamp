```python
df['Avarage'] = (df['Open']+df['Close'])/2
df.head(3)
```

```python
df = df.sort_index()
```

```python
df
```

```python
# dzienna zmiana procentowa ceny close

df['Close_shift'] = df[['Close']].shift(1) #przesuniÄ™cie o jednÄ… wartoÅ›Ä‡ w dÃ³Å‚
df.head(3)
```

```python
df['Daily_change'] = df['Close']/df['Close_shift']-1
df.head(3)
```
